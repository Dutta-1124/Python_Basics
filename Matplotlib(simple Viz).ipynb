```python
x=np.arange(0,4*np.pi,0.1)
y=np.sin(x)
```

```python
plt.title("Sin wave form")
plt.plot(x,y)
```

```python
#subplots

x=np.arange(0,5*np.pi,0.1)
y_sin=np.sin(x)
y_cos=np.cos(x)


plt.subplot(2,1,1)
plt.plot(x,y_sin,"r--")
plt.title("Sine")

plt.subplot(2,1,2)
plt.plot(x,y_cos,"g--")
plt.title("Cosine")
```

```python
#Barplot
x=[2,8,10]
y=[11,16,9]

x2=[3,9,11]
y2=[6,15,7]

plt.bar(x,y)
plt.bar(x2,y2,color="g")

plt.title("Bar graph")
plt.ylabel("Y axis")
plt.xlabel("X axis")
```

```python
#histograms
a = np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27]) 
plt.hist(a)
plt.title("Histogram")
#default bins=10
```

```python
#Piechart
```

```python
labels = 'Python', 'C++', 'Ruby', 'Java'
sizes = [215, 130, 245, 210]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.4, 0, 0, 0)  # explode 1st slice

# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=False)

plt.axis('equal')
plt.show()
```
